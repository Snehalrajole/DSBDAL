```python
import pandas as pd
import numpy as np
df=pd.read_csv('StudentsPerformanceTest1 (1).csv')
df
```

```python
#2.Checknull valuesinthedataset
print("Nullvalues in eachcolumn:")
print(df.isnull())
```

```python
#3.Countthenumberofnullvaluesincompletedataset
 #Replace missingvalueslikeblankstringsor'NA'withstandardNaN
df.replace(['','NA', 'N/A', 'na', '--'],np.nan,inplace=True)
null_count= df.isnull().sum().sum()
print("\nTotal number of null values in the dataset:",null_count)
```

```python
 #4.Droppingrowswith atleast1nullvalue
 df_dropped_any= df.dropna()
 print("\nDataFrame after dropping rows with at least 1 null value:")
 print(df_dropped_any)
```

```python
#5.Droppingrowsifallvaluesinthatrowaremissing
df_dropped_all= df.dropna(how='all')
print("\nDataFrame after dropping rows where all values are null:")
print(df_dropped_all)
```

```python
 #7.Droppingrowswith atleast1nullvalueandsavingtonewCSV
df_cleaned= df.dropna()
df_cleaned.to_csv('academic_performance_cleaned.csv',index=False)
print("\nRowswith null values dropped and saved to'academic_performance_cleaned.csv'")
```
