```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

df=pd.read_csv('/content/demo.csv')
df
```

```python
# Step 2: Sort Reading Score feature
sorted_rscore = sorted(df['reading score'])

print("Sorted Reading Scores:")
print(sorted_rscore)
```

```python
# Step 4: Calculate and print Quartile 1 and Quartile 3
q1 = df['reading score'].quantile(0.25)
q3 = df['reading score'].quantile(0.75)
print("Q1:", q1)
print("Q3:", q3)
```

```python
# Step 5: Calculate value of IQR
IQR = q3 - q1

# Step 6: Calculate and print Upper and Lower Bound
lwr_bound = q1 - (1.5 * IQR)
upr_bound = q3 + (1.5 * IQR)
print("Lower Bound:", lwr_bound)
print("Upper Bound:", upr_bound)
```

```python
# Step 7: Print Outliers
r_outliers = []
for i in sorted_rscore:
    if i < lwr_bound or i > upr_bound:
        r_outliers.append(i)

print("Outliers in Reading Score:")
print(r_outliers)
```

```python
# Step 8: Remove Outliers from the DataFrame
# Filter out rows where the 'reading score' column contains outliers
df_clean = df[(df['reading score'] >= lwr_bound) & (df['reading score'] <= upr_bound)]


# Step 9: Print the cleaned dataset size and compare with original size
print("\nOriginal Dataset Size:", df.shape)
print("Cleaned Dataset Size:", df_clean.shape)
df_clean
```
