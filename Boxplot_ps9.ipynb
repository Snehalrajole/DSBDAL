```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("/content/demo.csv")  # Update path based on your environment
df
```

```python
# Select the columns for the BoxPlot and draw the plot
col = ['math score', 'reading score', 'writing score', 'placement score']
df.boxplot(col)
```

```python
print(np.where(df['math score'] > 90))  # Adjust threshold as per data
print(np.where(df['math score'] < 60))  # Adjust threshold as per data
```

```python
#flooring and Capping
ninetieth_percentile = np.percentile(df['math score'], 90)  # Calculate 90th percentile
tenth_percentile = np.percentile(df['math score'], 10)  # Calculate 10th percentile

print(ninetieth_percentile)
print(tenth_percentile)

# Apply flooring and capping
df['Updated math score'] = np.where(df['math score'] > ninetieth_percentile,
                            ninetieth_percentile, df['math score'])
df['Updated math score'] = np.where(df['Updated math score'] < tenth_percentile,
                            tenth_percentile, df['Updated math score'])

print("New Updated column: ", np.array(df['Updated math score']))


# Insert the new column with capped and floored values (as in PDF)
#df.insert(1, "Updated Math Score", df['math score'], True)

print(df)
```
