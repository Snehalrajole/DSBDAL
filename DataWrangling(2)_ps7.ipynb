```python
#1. Load the Academic Performance dataset in data frame object.
#2. Check null values in the dataset.
#3. Check missing values in dataset  and replace the null values with standard null value NaN
#4.  Replace the missing  value of Math Score with Mean Value
#5.  Replace the missing value of Reading Score with standard deviation
#6.  Replace the missing value of place with common value "Nashik"
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/StudentsPerformanceTest1.csv')
df
```

```python
df.isnull()
# Step 2: Check for null values in the dataset
```

```python
df.isnull().sum()
```

```python
df.replace('', np.nan, inplace=True)
df
# Step 3: Standardize missing values (if not already None or NaN)
```

```python
#math_mean = df['math score'].mean()
#df['math score'].fillna(math_mean, inplace=True)
# Step 4: Replace missing values in 'math_score' with mean#



df['math score'] = pd.to_numeric(df['math score'], errors='coerce')

# Calculate the mean of 'math score'
math_mean = df['math score'].mean()

# Replace missing values with the mean
df['math score'].fillna(math_mean, inplace=True)

print("Updated 'math score' column:\n", df['math score'])
```

```python
# Step 5: Replace missing values in 'reading_score' with standard deviation
#reading_std = df['reading score'].std()
#df['reading score'].fillna(reading_std, inplace=True)


 #Convert 'reading score' to numeric, coercing errors to NaN
df['reading score'] = pd.to_numeric(df['reading score'], errors='coerce')

# Calculate the standard deviation of 'reading score'
reading_std = df['reading score'].std()

# Replace missing values with the standard deviation
df['reading score'].fillna(reading_std, inplace=True)

print("Updated 'reading score' column:\n", df['reading score'])
```

```python
# Step 6: Replace missing values in 'place' with the common value "Nashik"
df['Region'].fillna('Nashik', inplace=True)
df
```
